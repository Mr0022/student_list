```python
import pandas as pd
import pprint
import json

# Loading data and converting keys to integers
def load_data():
    try:
        with open('data.json', 'r') as json_file:
            loaded_data = json.load(json_file)
            # Convert keys back to integers
            return {int(k): v for k, v in loaded_data.items()}
    except FileNotFoundError:
        return {}

data = load_data()
# validating if the student_id input is correct

def validate_input_Id(prompt, condition=lambda x: True, error_msg="Invalid input"):
    while True:
        try:
            value = int(input(prompt))
            if condition(value):
                return value
            else:
                print(error_msg)
        except ValueError:
            print(error_msg)

# validating if the student_score input is correct

def validate_input_score(prompt, condition=lambda x: True, error_msg="Invalid input"):
    while True:
        try:
            user_input = input(prompt)
            if user_input == "":
                return None  
            value = int(user_input)
            if condition(value):
                return value
            else:
                print(error_msg)
        except ValueError:
            print(error_msg)

# validating if the student first name,lastname input is correct and have at least 3 character

def validate_input_char(prompt, error_msg="Invalid input"):
    while True:
        user_input = input(prompt)
        if user_input == "":
            return None 
        if len(user_input) > 2 :
            return user_input
        else:
            print(error_msg)



def add_student():
    Student_Id = validate_input_Id("Enter new student ID : ", lambda x: x not in data, "Student ID already exists or invalid.")
    firstname = validate_input_char("Enter First Name: "," At least 3 character is necessary")
    lastname = validate_input_char("Enter Last Name: ","At least 3 character is necessary" )
    score = validate_input_score("Enter Score: ", lambda x: 0 <= x <= 20, "Score must be between 0 and 20.")
    data[Student_Id] = {"firstname": firstname, "lastname": lastname, "score": score}
    return "Student added successfully."

def delete_student():
    Student_Id = validate_input_Id("Enter student ID to delete: ", lambda x: x in data, "Student ID not found or invalid.")
    del data[Student_Id]
    return "Student deleted successfully."

def edit_student():
    Student_Id = validate_input_Id("Enter student ID to edit: ", lambda x: x in data, "Student ID not found or invalid.")
    firstname = validate_input_char("Enter First Name (leave blank to keep current): ")
    lastname = validate_input_char("Enter Last Name (leave blank to keep current): ")
    score = validate_input_score("Enter Score (leave blank to keep current): ", lambda x: x is None or 0 <= x <= 20, "Score must be between 0 and 20.")
    if firstname:
        data[Student_Id]['firstname'] = firstname
    if lastname:
        data[Student_Id]['lastname'] = lastname
    if score:
        data[Student_Id]['score'] = int(score)
    return "Student updated successfully."

# this function to show data as a dataset
def show_data():
    df = pd.DataFrame.from_dict(data, orient='index')
    print(df)


# Save data to JSON file
def save_data():
    with open('data.json', 'w') as json_file:
        json.dump(data, json_file)


# Interactive menu
def menu():
    while True:
        print("\nMenu")
        print("1. Add New Student")
        print("2. Delete Existing Student")
        print("3. Edit Student Specification")
        print("4. Show Data")
        print("5. Exit")
        choice = input("Just enter the number from the list: ")
        
        if choice == '1':
            print(add_student())
        elif choice == '2':
            print(delete_student())
        elif choice == '3':
            print(edit_student())
        elif choice == '4':
            #show_data()       # we can use this function instead of below to show data in dataset
            pprint.pprint(data)
        elif choice == '5':
            save_data()
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

# Run the menu
menu()
```
